```python
import pandas as pd
data = pd.read_csv("12_means_clustering_final_predict.csv")
```

```python
cluster = {}
for i in range(12):
    cluster[i] = []
for i in range(12):
    cluster[i] = data[data['cluster']==i]['wbt_id'].tolist()
cluster
```

```python
import pickle

with open('cluster_12_final_predict.pickle', 'wb') as fw:
    pickle.dump(cluster, fw)
    
with open('cluster_12_final_predict.pickle', 'rb') as fr:
    cluster_loaded = pickle.load(fr)

print(cluster_loaded)
```
