```python
import pandas as pd
```

```python
Ndata = pd.read_csv('/content/drive/MyDrive/webtoon/N_Webtoon_1_toonsinfo.csv')
Ddata = pd.read_csv('/content/drive/MyDrive/webtoon/D_Webtoon_1_toonsinfo.csv')
```

```python
Ndata.drop(['Unnamed: 0','finishYn','artist','latest_upload','img_url','intro','form'],axis = 1, inplace = True)
Ndata.head(5)
```

```python
Ddata.drop(['nickname','finishYn','ageGrade','dateCreated','artist','latest_upload','last_chapter','img_url','categories','intro'],axis = 1, inplace = True)
Ddata.head(5)
```

```python
Ndata=Ndata[['id','title','Score','page url','genres']]
Ndata.head(5)
```

```python
Ddata.columns = ['id', 'title', 'Score', 'page url', 'genres']
Ddata.head()
```

```python
Ddata['genres'] = Ddata['genres'].str.replace('/',' ')
Ddata.head()
```

```python
Ddata['genres']= Ddata['genres'].astype(str).str.split()
type(Ddata['genres'][2])
```

```python
Ddata.head()
```

```python
new_train_df = Ddata.copy()
new_train_df.head()
```

```python
result = new_train_df['genres']
result
```

```python
result = result.apply(lambda x: pd.Series(x))
result.stack()
result = result.stack().reset_index(level=1, drop=True).to_frame('single_genre')
result
```

```python
result = new_train_df.merge(result, left_index=True, right_index=True, how='left')
result
```

```python
result.drop(['genres'],axis=1, inplace=True)
result.columns = ['id', 'title', 'Score', 'page url', 'genres']
result.head()
```

```python
Ndata['genres'] = Ndata['genres'].str.replace('daily','일상')
Ndata['genres'] = Ndata['genres'].str.replace('comic','코믹')
Ndata['genres'] = Ndata['genres'].str.replace('fantasy','판타지')
Ndata['genres'] = Ndata['genres'].str.replace('action','액션')
Ndata['genres'] = Ndata['genres'].str.replace('drama','드라마')
Ndata['genres'] = Ndata['genres'].str.replace('pure','순정')
Ndata['genres'] = Ndata['genres'].str.replace('thrill','스릴러')
Ndata['genres'] = Ndata['genres'].str.replace('sports','스포츠')
Ndata['genres'] = Ndata['genres'].str.replace('historical','시대극')
Ndata['genres'] = Ndata['genres'].str.replace('sensibility','감성')

Ndata
```

```python
total_data = pd.concat([Ndata,result])
total_data
```

```python
train_data=total_data.copy()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
encoder.fit(train_data['genres'])
train_data['genres'] = encoder.transform(train_data['genres'])
train_data
```

```python
mapping = dict(zip(range(len(encoder.classes_)), encoder.classes_))
mapping
```

```python
collect_number=[]
```

```python
def pick_genre_number(df,webtoon_title):
    target_webtoon = df[df['title']==webtoon_title]
    genre_number = int(target_webtoon['genres'])
    collect_number.append(genre_number)
```

```python
pick_genre_number(train_data,'프리스타일 풋볼')
pick_genre_number(train_data,'독립일기')
pick_genre_number(train_data,'오빠세끼')
pick_genre_number(train_data,'군대라면')
pick_genre_number(train_data,'윌유메리미')
collect_number
```

```python
def get_recommend_webtoon_list(df, collect_number):
    count={}
    for i in collect_number:
        try: count[i] += 1
        except: count[i]=1
    lists = [k for k,v in count.items() if max(count.values()) == v]
    genre = ['당신의 선호 장르는']
    for j in lists:
        genre.append(mapping.get(j))
    genre.append('입니다.')
    print(" ".join(genre))
    result = df[df['genres'].isin(lists)].sort_values('Score',ascending = False)
    result.drop(['genres'],axis=1, inplace=True)
    return result
```

```python
get_recommend_webtoon_list(train_data,collect_number)
```
